## csv mean and std columns

```python
import os
import cv2
import numpy as np
import pandas as pd

# Path to your folder
folder_path = r"D:\Aamir Gulzar\KSA_project2\Cancer-detection-classifier\Data_cleaning\Data"

# Initialize a list to hold data
data = []

# Iterate over all PNG files in the folder
for filename in os.listdir(folder_path):
    if filename.lower().endswith('.png'):
        filepath = os.path.join(folder_path, filename)
        
        # Read the image using OpenCV (in BGR format)
        image = cv2.imread(filepath)

        # Convert BGR to RGB
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Compute mean and std for each channel
        mean_r = np.mean(image_rgb[:, :, 0])
        mean_g = np.mean(image_rgb[:, :, 1])
        mean_b = np.mean(image_rgb[:, :, 2])
        std_r = np.std(image_rgb[:, :, 0])
        std_g = np.std(image_rgb[:, :, 1])
        std_b = np.std(image_rgb[:, :, 2])

        # Append to data list
        data.append({
            'patch_name': filename,
            'avg_R': mean_r,
            'avg_G': mean_g,
            'avg_B': mean_b,
            'std_R': std_r,
            'std_G': std_g,
            'std_B': std_b
        })
```

```python
# Convert to DataFrame
df = pd.DataFrame(data)
df
```
